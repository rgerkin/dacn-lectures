```python
# This is the main plotting library we will use
import matplotlib.pyplot as plt
# And this is the pandas library we have been using for handling data
import pandas as pd
import numpy as np
```

```python
# Make up some fake data
data = np.random.rand(100,4)+np.sin(2*np.pi*np.arange(100)/20)[:,np.newaxis]
```

```python
# Let's load a simple dataset and use the date as the index
df = pd.DataFrame(data,columns=['A','B','C','D'])
df.index.name = 'Time'
df.head(10)
```

## Let's look at other plot types

```python
# Font sizes are usually way too small!  Let's increase them.  
import matplotlib as mpl
mpl.rcParams.update({'font.size':16})
```

```python
# Line plot
df.plot(ylim=(-2,6));
```

```python
# Line plot
df.plot.scatter('A','B',s=50);
```

```python
# Bar plot
df.iloc[:10].plot.bar(ylim=(0,3),rot=45)
```

```python
# Bubble chart
df.plot.scatter('A','B',s=30*df['C']**2);
```

```python
# Color Bubble chart
df.plot.scatter('A','B',s=30*df['C']**2,c=df['D']);
```

```python
df.plot.hist(subplots=True,layout=(2,2),figsize=(7,5));
plt.tight_layout()
```

```python
# Heatmap
plt.pcolor(df)
plt.xticks(np.arange(0.5,4.5),df.columns);
plt.colorbar(label='Value')
```

```python
# Joint histogram
df.plot.hexbin('A','B',gridsize=10);
plt.tight_layout()
```

```python
# Pie chart
df['E'] = 1*(df['A']>df['B'])
df['E'] += df['C']>df['B']
df['E'].value_counts().plot.pie()
```
